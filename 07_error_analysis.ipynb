```python
import sys
sys.path.append("..")

import pandas as pd
from src.scoutai_common import get_engine, load_master_view, engineer_features, route_predict

engine = get_engine()
df = load_master_view(engine)
df = engineer_features(df)
df = route_predict(df)
df['prediction_error'] = abs(df['current_market_value'] - df['predicted_value'])

pd.options.display.float_format = '{:,.0f}'.format
columns_to_show = ['player_name', 'age', 'model_used', 'current_market_value', 'predicted_value', 'prediction_error']
rename_map = {'player_name': 'Player Name', 'age': 'Age', 'model_used': 'Model',
              'current_market_value': 'Actual Value (E)', 'predicted_value': 'ScoutAI Value (E)',
              'prediction_error': 'Gap (Error E)'}

for label in ['full', 'performance_only']:
    subset = df[df['model_used'] == label].sort_values(by='prediction_error', ascending=False)
    presentation_df = subset[columns_to_show].rename(columns=rename_map)
    print(f"\nTop 10 Most Confusing Players -- {label} model:")
    print("-" * 95)
    print(presentation_df.head(10).to_string(index=False))
```
